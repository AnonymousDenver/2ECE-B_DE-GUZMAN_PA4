```python
import pandas as np # Importing Liblary for pandas
import matplotlib.pyplot as plt #Importing liblary for matplotlib
import seaborn as sns # Importing Liblary for seaborn
```

```python
df = pd.read_excel('board2.xlsx') # load the file 
df 
```

```python
# Calculate the average and store it in a new column 'avg'
df['avg'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1) #axis 1 to get the mean by columns

# Print the average grades
print(df['avg'])
```

```python
#get the average for the different features (Gender, Track, Hometown)
subAvg = df.groupby(['Gender','Track','Hometown'])['avg'].mean()
subAvg
```

```python
#with the use of data visualization with seaborn and matplotlib, plot the bar graph of the that will show the contribution of gender in gaining a high average
colors = ['#FFB6C1','#AA336A']
plt.figure(figsize=(11, 6))
sns.barplot(x='Gender', y='avg', data=df, hue='Gender', palette=colors)
sns.set(style="whitegrid")
plt.show()
```

```python
#with the use of data visualization with seaborn and matplotlib, plot the bar graph of the that will show the contribution of track in college in gaining a high average
colors = ['#DAA520','#FFA07A','#FFDAB9'] # Color codes for phyton
plt.figure(figsize=(11, 4))
sns.barplot(x='Track', y='avg', data=df, hue='Track', palette=colors)
sns.set(style="whitegrid")
plt.show()
```

```python
#with the use of data visualization with seaborn and matplotlib, plot the bar graph of the that will show the contribution of hometown in gaining a high average
colors = ['#FF00FF','#CC00CC','#990099']
plt.figure(figsize=(10, 4))
sns.barplot(x='Hometown', y='avg', data=df, hue='Hometown', palette=colors)
sns.set(style="whitegrid")
plt.show()
```
